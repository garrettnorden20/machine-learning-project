```python
# Random Forest
allstats = allstats_corr._get_numeric_data()
allstats = allstats.fillna(0)

y = allstats['win']
X = allstats.drop(['win'], axis=1)

scaled_features = sci_pre.StandardScaler().fit_transform(X.values)
X = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
X_train, y_train, X_test, y_test = np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test)

```

```python

################# CHANGE THESE VALUES ###############################
n_est = 10 #Hint: Consider values between 5-12.
depth = 12 # Hint: Consider values betweeen 3-12.
max_feat = 0.9 # Hint: Consider values betweeen 0.6-1.0.
#####################################################################

#random_forest = randomforest.RandomForest(n_estimators, max_depth, max_features, random_seed=student_random_seed)
random_forest = sci_en.RandomForestClassifier(criterion='entropy', n_estimators=n_est, max_depth=depth, max_features = max_feat, random_state=0)
random_forest.fit(X_train, y_train)
    
accuracy=random_forest.score(X_test, y_test)
    
print("accuracy: %.4f" % accuracy)
```
