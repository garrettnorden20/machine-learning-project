```python
#df_corr = df._get_numeric_data()
allstats_reduced = allstats[['id', 'matchid', 'player', 'name', 'adjposition', 'team_role', 'win', 'kills', 'deaths', 'assists', 'turretkills','totdmgtochamp', 'totheal', 'totminionskilled', 'goldspent', 'totdmgtaken', 'inhibkills', 'pinksbought', 'wardsplaced', 'duration', 'platformid', 'seasonid', 'goldearned', 'version']]

plt.style.use('ggplot')
allstats_corr = allstats_reduced._get_numeric_data()
allstats_corr = allstats_corr.drop(['id', 'matchid', 'player', 'seasonid'], axis = 1)

mask = np.zeros_like(allstats_corr.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(10, 150, as_cmap=True)

plt.figure(figsize = (15,10))
sns.heatmap(allstats_corr.corr(), cmap = cmap, annot = True, fmt = '.2f', mask = mask, square=True, linewidths=.5, center = 0)
plt.title('Correlations - win vs factors (all games)')

# this 
```

```python

#print(X)
#print(y)
allstats = allstats_corr
allstats = allstats.fillna(0)

y = allstats['win']
X = allstats.drop(['win'], axis=1)

# scaling?
X = sci_pre.StandardScaler().fit_transform(X, y)

maxK = allstats.shape[1]

k_graph = np.arange(1, maxK)
accuracies = np.zeros((0))
accuracies_pca = np.zeros((0))


for k_iter in range(1,maxK):
    # for select K best
    
    #X_ = sci_feat.SelectKBest(sci_feat.chi2, k=k_iter).fit_transform(X, y)
    X_ = sci_feat.SelectKBest(sci_feat.f_regression, k=k_iter).fit_transform(X, y)
    accuracy = ml.logistic_regression(X_,y)
    accuracies = np.append(accuracies, accuracy)
    print('KBest', k_iter, accuracy)
    
    
    # for PCA
    X_ = sci_de.PCA(k_iter).fit_transform(X, y)
    accuracy = ml.logistic_regression(X_,y)
    accuracies_pca = np.append(accuracies_pca, accuracy)
    print('PCA', k_iter, accuracy)

plt.plot(k_graph, accuracies)
plt.plot(k_graph, accuracies_pca)
plt.show()
```

```python
# PCA elbow explained variance
pca = sci_de.PCA(maxK-1)
X_ = pca.fit_transform(X, y)

k_graph = np.arange(1, maxK)
print(pca.explained_variance_ratio_)
#plt.plot(k_graph, pca.explained_variance_ratio_)
plt.plot(k_graph, np.cumsum(pca.explained_variance_ratio_))
```
